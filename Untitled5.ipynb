```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import pandas as pd

# 図やグラフを図示するためのライブラリをインポートする。
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn #機械学習のライブラリ
from sklearn.decomposition import PCA #主成分分析器
```

```python
dataset_train=pd.read_csv('PM_train.txt',sep=' ',header=None).drop([26,27],axis=1)
col_names = ['id','cycle','setting1','setting2','setting3','s1','s2','s3','s4','s5','s6','s7','s8','s9','s10','s11','s12','s13','s14','s15','s16','s17','s18','s19','s20','s21']
dataset_train.columns=col_names
print('Shape of Train dataset: ',dataset_train.shape)
dataset_train.head()
```

```python
# 説明変数
X_test = dataset_train.drop(['id','cycle'],1)
```

```python
# 行列の標準化
X_tests = X_test.iloc[:, 1:].apply(lambda x: (x-x.mean())/x.std(), axis=0)
X_tests.head()
```

```python
#主成分分析の実行
pca = PCA()
feature = pca.fit(X_tests)
# データを主成分空間に写像
feature = pca.transform(X_tests)
```
