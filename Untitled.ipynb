```python
dataset_train=pd.read_csv('PM_train.txt',sep=' ',header=None).drop([26,27],axis=1)
col_names = ['id','cycle','setting1','setting2','setting3','s1','s2','s3','s4','s5','s6','s7','s8','s9','s10','s11','s12','s13','s14','s15','s16','s17','s18','s19','s20','s21']
dataset_train.columns=col_names
print('Shape of Train dataset: ',dataset_train.shape)
dataset_train.head()
```

```python
import numpy as np
from pandas import Series, DataFrame
import pandas as pd

dataset_train = DataFrame(dataset_train,columns=['id','label1','label2','cycle','setting1','setting2','setting3','s1','s2','s3','s4','s5','s6','s7','s8','s9','s10','s11','s12','s13','s14','s15','s16','s17','s18','s19','s20','s21'])
```

```python
i_max_temp = 0
```

```python
for i in range (1,3):

    i_dataset_train = dataset_train[dataset_train.id > (i - 1)]
    i_dataset_train = i_dataset_train[dataset_train.id < (i + 1)]


    i_max = i_dataset_train['cycle'].max()

    for n in range (i_max_temp, i_max_temp + i_max - 30):
        dataset_train.iat[n, 1] = 0

    for n in range (i_max_temp + i_max - 30, i_max):
        dataset_train.iat[n, 1] = 1
        
    i_max_temp = i_max_temp + i_max
```
