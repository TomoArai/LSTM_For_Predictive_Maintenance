```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import pandas as pd
```

```python
dataset_train=pd.read_csv('PM_train.txt',sep=' ',header=None).drop([26,27],axis=1)
col_names = ['id','cycle','setting1','setting2','setting3','s1','s2','s3','s4','s5','s6','s7','s8','s9','s10','s11','s12','s13','s14','s15','s16','s17','s18','s19','s20','s21']
dataset_train.columns=col_names
print('Shape of Train dataset: ',dataset_train.shape)
dataset_train.head()
```

```python
dataset_train.head(200)
```

```python
dataset_train['ttf'] = dataset_train.groupby(['id'])['cycle'].transform(max)-dataset_train['cycle']
dataset_train.head()
```

```python
dataset_train['label1'] = dataset_train.groupby(['id'])['cycle'].transform(max)-dataset_train['cycle']
dataset_train.head()
```

```python
for i in range(20631):
    if dataset_train.at[i, 'label1'] > 29:
        dataset_train.at[i, 'label1'] = 0
    elif dataset_train.at[i, 'label1'] <= 29 and dataset_train.at[i, 'label1'] >14 :
        dataset_train.at[i, 'label1'] = 1
    else:
        dataset_train.at[i, 'label1'] = 2
```

```python
dataset_train =  dataset_train.drop(['cycle'],1)
dataset_train.head()
```

```python
# 説明変数
X_multi = dataset_train.drop(['ttf','label1'],1)

# 目的変数
Y_target = dataset_train.label1
```

```python
X_multi.head()
```

```python
Y_target.head()
```

```python
import sklearn
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split

# 精度を計算するのに便利なツール
from sklearn import metrics

# おなじく、train_test_splitを使う
X_train, X_test, Y_train, Y_test = train_test_split(X_multi, Y_target, test_size=0.3)
```

```python
# 決定木
from sklearn.tree import DecisionTreeClassifier

# 決定木インスタンス(木の深さ3)
tree_model = DecisionTreeClassifier(max_depth=3)
#学習モデル構築。引数に訓練データの特徴量と、それに対応したラベル
tree_model.fit(X_train, Y_train)

# .scoreで正解率を算出。
print("train score:",tree_model.score(X_train,Y_train))
print("test score:",tree_model.score(X_test,Y_test))
```
